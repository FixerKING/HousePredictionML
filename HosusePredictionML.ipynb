### Feature Engineering and Model Selection

```python
features = ['LotArea', '1stFlrSF', '2ndFlrSF', 'YearBuilt', 'TotRmsAbvGrd', 'FullBath', 'BedroomAbvGr']
```

```python
y = data_train.SalePrice
X = data_train[features].copy()
X_test = data_test[features].copy()
```

### Model Definition

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
model_1 = RandomForestRegressor(n_estimators = 50, random_state = 0)
model_2 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute error', random_state = 0)
model_3 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_4 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

models = [model_1, model_2, model_3, model_4]
```

```python
model_3.fit(X,y)
preds = model_3.predict(X_valid)
error = mean_absolute_error(y_valid, preds)
print(error)
```
